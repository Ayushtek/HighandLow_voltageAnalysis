```python
import pandas as pd
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.graph_objs as go

 # Convert timestamp to datetime if needed

# Initialize the Dash app
app = dash.Dash(__name__)

# Layout of the dashboard
app.layout = html.Div([
    html.H1("Madhu Vihar DT1 Voltage and Current Dashboard"),
    
    # Dropdown for selecting the parameter (Y, R, B, N)
    html.Label("Select Parameter:"),
    dcc.Dropdown(
        id='parameter-dropdown',
        options=[
            {'label': 'Y', 'value': 'Y'},
            {'label': 'R', 'value': 'R'},
            {'label': 'B', 'value': 'B'},
            {'label': 'N', 'value': 'N'}
        ],
        value='Y'
    ),
    
    # Date-time range picker
    html.Label("Select Date-Time Range:"),
    dcc.DatePickerRange(
        id='date-picker',
        start_date=df['timestamp'].min().date(),
        end_date=df['timestamp'].max().date(),
        display_format='YYYY-MM-DD'
    ),
    
    # Graph output
    dcc.Graph(id='time-series-graph')
])

# Callback for updating the graph
@app.callback(
    Output('time-series-graph', 'figure'),
    [Input('parameter-dropdown', 'value'),
     Input('date-picker', 'start_date'),
     Input('date-picker', 'end_date')]
)
def update_graph(selected_param, start_date, end_date):
    # Filter data by date range
    filtered_data = df[(df['timestamp'] >= start_date) & (df['timestamp'] <= end_date)]
    
    # Create the graph
    figure = go.Figure()
    figure.add_trace(go.Scatter(
        x=filtered_data['timestamp'],
        y=filtered_data[selected_param],
        mode='lines',
        name=selected_param
    ))
    
    figure.update_layout(
        title=f"{selected_param} vs Timestamp",
        xaxis_title="Timestamp",
        yaxis_title=selected_param,
        template="plotly_dark"
    )
    return figure

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True, port=9000)
```

```python
def classify_alert(phase_diff):
    if phase_diff > 265:
        return 'High Alert'
    elif phase_diff < 216:
        return 'Low Alert'
    elif 216 <= phase_diff <= 235:
        return 'Low Warning'
    elif 235 < phase_diff <= 265:
        return 'Normal'
    return 'Unknown'

# Apply the classify_alert function to each phase difference column
n1['phaserangeR'] = n1['phasediffR'].apply(classify_alert)
n1['phaserangeY'] = n1['phasediffY'].apply(classify_alert)
n1['phaserangeB'] = n1['phasediffB'].apply(classify_alert)


n1
```

## MONTHLY-WISE OCCURENCES 

```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Assuming `low_warning_df` is your dataframe
# Step 1: Ensure 'record_time' is in datetime format and extract the month
low_warning_df['record_time'] = pd.to_datetime(low_warning_df['record_time'])
low_warning_df['month'] = low_warning_df['record_time'].dt.to_period('M')

# Step 2: Round the 'time_diff_minutes' column
low_warning_df['rounded_time_diff_minutes'] = low_warning_df['time_diff_minutes'].round()

# Step 3: Create a subplot for each unique month (left to right)
months = low_warning_df['month'].unique()
cols = len(months)  # Number of columns is the number of unique months
fig = make_subplots(
    rows=1, cols=cols, 
    shared_yaxes=True, 
    horizontal_spacing=0.1,
    subplot_titles=[str(month) for month in months]
)

# Step 4: Group the data by month and plot the count of occurrences for each rounded time difference
for i, month in enumerate(months, start=1):
    # Filter data for the current month
    month_data = low_warning_df[low_warning_df['month'] == month]
    
    # Group by the rounded time difference and count the occurrences
    time_diff_counts = month_data.groupby('rounded_time_diff_minutes').size().reset_index(name='count')
    
    # Create the bar chart for this month
    bar = go.Bar(
        x=time_diff_counts['rounded_time_diff_minutes'],
        y=time_diff_counts['count'],
        text=time_diff_counts['count'],  # Display count on top of bars
        textposition='auto'
    )
    
    # Add the bar chart to the corresponding subplot
    fig.add_trace(bar, row=1, col=i)

# Step 5: Customize layout
fig.update_layout(
    title="Count of Occurrences by Rounded Time Difference (Minutes) for Each Month",
    xaxis_title="Time Difference (Minutes)",
    yaxis_title="Count of Occurrences",
    template="plotly",  # Optional: Change the theme if you prefer a different style
    showlegend=False,
    xaxis=dict(
        tickmode='linear',  # Mark every minute on the x-axis
        dtick=1,            # 1-minute intervals
    ),
    height=600,  # Optional: Adjust height to fit all subplots well
)

# Step 6: Show the interactive plot
fig.show(renderer='browser')
```

```python
import pandas as pd
import plotly.express as px

# Assuming 'n_high1' is your dataset

# Filter the rows where any of the phase ranges is 'High Alert'
filtered_df = n_high1[(n_high1['phaserangeY'] == 'High Alert') | 
                      (n_high1['phaserangeR'] == 'High Alert') | 
                      (n_high1['phaserangeB'] == 'High Alert')]

# Extract the month and hour of day from 'record_time'
filtered_df['record_time'] = pd.to_datetime(filtered_df['record_time'])  # Make sure it's in datetime format
filtered_df['month'] = filtered_df['record_time'].dt.month
filtered_df['hour_of_day'] = filtered_df['record_time'].dt.hour

# Filter data to only include records from November 2023 to February 2024
filtered_df = filtered_df[(filtered_df['record_time'].dt.month >= 11) & 
                          (filtered_df['record_time'].dt.year == 2023) | 
                          (filtered_df['record_time'].dt.year == 2024) & 
                          (filtered_df['record_time'].dt.month <= 2)]

# Group by month and hour to count unique instances of 'High Alert'
hourly_instances_by_month = filtered_df.groupby(['month', 'hour_of_day']).size().reset_index(name='count')

# Group by month to calculate total hours of high alert warnings for each month
monthly_total_hours = hourly_instances_by_month.groupby('month')['count'].sum().reset_index(name='total_hours')

# Create a bar plot to visualize the total high alert hours per month
fig = px.bar(monthly_total_hours, 
             x='month', 
             y='total_hours',
             title='Month-wise Total Hours of High Voltage Warnings (Nov 2023 - Feb 2024)',
             labels={'month': 'Month', 'total_hours': 'Total High Alert Hours'},
             color='total_hours',  # Color bars based on total hours
             color_continuous_scale='Viridis')  # Color scale for the bars

# Update layout for better readability
fig.update_layout(
    xaxis=dict(tickmode='array', tickvals=[11, 12, 1, 2], ticktext=["Nov 2023", "Dec 2023", "Jan 2024", "Feb 2024"]),
    template='plotly_white',  # Optional: White background theme for better visibility
    height=500,  # Adjust the height for better readability
    width=800,  # Adjust the width for a better fit
    hovermode="x unified",  # Unified hover effect
    showlegend=False  # Hide legend since it's not needed
)

# Show the plot in the browser
fig.show(renderer='browser')
```
